```python
import pandas as pd
import shutil
import os
import wfdb
```

```python
file = pd.read_csv(os.path.join('../../data/ptbxl/ptbxl_database.csv'))
file = file[0:10]

for idx, row in file.iterrows():
    print(f"Processing record {row['ecg_id']} from patient {row['patient_id']}")
    file_name = f"../../data/ptbxl/{row['filename_lr']}"
    wfdb.rdsamp(file_name)
    dest_file = f"./{row['filename_lr']}"
    shutil.copy(file_name + ".dat", dest_file + ".dat")
    shutil.copy(file_name + ".hea", dest_file + ".hea")
```
